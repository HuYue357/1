```python
# --- 参数区：按你的实际路径修改 ---
MASTER_CSV = "Origin/origin.csv"    # 第一列 query 名，第二列时间（毫秒）
COMBOS_DIR = "All"                  # 目录下 q1.csv … q99.csv
TIME_SCALE = 1.0                    # 若组合文件是秒，则设 1000.0

# =================================
import os, glob, re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -------- 名称归一化：统一成 q<int>，容错大小写/空格/前导0/尾缀 --------
def norm_q(s: str) -> str:
    if s is None:
        return None
    s = str(s).strip().lower()
    # 有 q 前缀
    m = re.search(r'q\s*0*([0-9]+)', s)
    if m:
        return f"q{int(m.group(1))}"
    # 纯数字也兼容
    m2 = re.search(r'0*([0-9]+)', s)
    return f"q{int(m2.group(1))}" if m2 else s

# -------- 读取原版 Spark 耗时 --------
def load_master(path):
    df = pd.read_csv(path)
    if df.shape[1] < 2:
        raise ValueError("MASTER_CSV 需要至少两列：[query, time_ms]")
    df = df.iloc[:, :2].copy()
    df.columns = ["query_raw", "default_ms"]
    df["query"] = df["query_raw"].map(norm_q)
    df["default_ms"] = pd.to_numeric(df["default_ms"], errors="coerce")
    df = df.dropna(subset=["query", "default_ms"])
    # 若同名多行，取最小（也可改为 mean）
    df = df.groupby("query", as_index=False)["default_ms"].min()
    return df

# -------- 读取单个查询的笛卡尔积组合结果 --------
def load_combo_times(csv_path):
    # 约定：第二列为组合耗时（若只有一列就用第一列）
    df = pd.read_csv(csv_path, header=None)
    if df.shape[1] < 2:
        times = pd.to_numeric(df.iloc[:, 0], errors="coerce")
    else:
        times = pd.to_numeric(df.iloc[:, 1], errors="coerce")
    times = times.dropna()
    times = (times * TIME_SCALE).astype(float)
    return times.values

def load_all_combos(combos_dir):
    rows = []
    for fp in glob.glob(os.path.join(combos_dir, "*.csv")):
        base = os.path.splitext(os.path.basename(fp))[0]  # 'q35'
        qname = norm_q(base)
        try:
            times = load_combo_times(fp)
            for i, t in enumerate(times):
                rows.append((qname, i, t))
        except Exception as e:
            print(f"[WARN] 读取失败: {fp} -> {e}")
    if not rows:
        raise RuntimeError("未在 COMBOS_DIR 读到任何组合耗时。请检查目录/文件格式。")
    df = pd.DataFrame(rows, columns=["query", "combo_id", "combo_ms"])
    return df

# -------- 主流程 --------
master = load_master(MASTER_CSV)
combos = load_all_combos(COMBOS_DIR)

# 合并前打印差集，便于定位不匹配
left_only = sorted(set(master["query"]) - set(combos["query"]))
right_only = sorted(set(combos["query"]) - set(master["query"]))
if left_only or right_only:
    print("[提示] 名称差集：")
    if left_only:
        print("  主CSV独有(前20): ", left_only[:20], "..." if len(left_only) > 20 else "")
    if right_only:
        print("  组合独有(前20):  ", right_only[:20], "..." if len(right_only) > 20 else "")

# 合并默认耗时
all_data = combos.merge(master, on="query", how="inner")
if all_data.empty:
    raise RuntimeError("组合数据与默认数据的查询名没有交集，请检查命名（例如 q1 vs Q01、空格/前缀等）。")

# 计算指标
all_data = all_data[all_data["combo_ms"] > 0].copy()       # 避免除零/无意义
all_data["speedup"] = all_data["default_ms"] / all_data["combo_ms"]   # >1 表示组合更快
all_data["faster"] = all_data["combo_ms"] < all_data["default_ms"]

# 每个查询的汇总
summary = (all_data
    .groupby("query")
    .agg(
        default_ms=("default_ms", "first"),
        n_combos=("combo_ms", "size"),
        share_faster=("faster", "mean"),
        best_ms=("combo_ms", "min"),
        best_speedup=("speedup", "max"),
        median_ms=("combo_ms", "median"),
        median_speedup=("speedup", "median"),
        p10_ms=("combo_ms", lambda x: np.percentile(x, 10)),
        p25_ms=("combo_ms", lambda x: np.percentile(x, 25)),
        p75_ms=("combo_ms", lambda x: np.percentile(x, 75)),
        p90_ms=("combo_ms", lambda x: np.percentile(x, 90)),
        mean_ms=("combo_ms", "mean")
    )
    .reset_index()
)

# 总览打印
overall_share_faster = all_data["faster"].mean()
n_queries = summary.shape[0]
n_queries_median_better = (summary["median_ms"] < summary["default_ms"]).sum()
n_queries_best20 = (summary["best_ms"] < summary["default_ms"] * 0.8).sum()

print("===== 全局总览 =====")
print(f"查询数: {n_queries}")
print(f"总组合数: {all_data.shape[0]}")
print(f"所有组合中优于默认的占比: {overall_share_faster:.2%}")
print(f"中位组合耗时优于默认的查询数: {n_queries_median_better}/{n_queries} "
      f"({n_queries_median_better/n_queries:.2%})")
print(f"存在 ≥20% 提升的查询数（最佳组合对默认）: {n_queries_best20}/{n_queries} "
      f"({n_queries_best20/n_queries:.2%})")

# 导出汇总
out_csv = "gnn_combo_vs_default_summary.csv"
summary.to_csv(out_csv, index=False)
print(f"[保存] 每查询汇总 -> {out_csv}")

# -------- 图1：所有组合速度提升比的 CDF --------
spd = all_data["speedup"].values
spd = spd[np.isfinite(spd) & (spd > 0)]
if spd.size > 0:
    xs = np.sort(spd)
    ys = np.arange(1, xs.size + 1) / xs.size
    plt.figure(figsize=(7, 5))
    plt.plot(xs, ys, drawstyle="steps-post")
    plt.xlabel("Speedup ratio (default_time / combo_time)")
    plt.ylabel("CDF")
    plt.title("CDF of Speedup over Default (All Combinations)")
    plt.grid(True, which="both", axis="both")
    plt.axvline(1.0, linestyle="--")  # 右侧为“更快”
    plt.tight_layout()
    plt.show()

# -------- 图2：每查询“更快组合占比”Top-30 柱状图 --------
top_share = (summary
             .sort_values(["share_faster", "n_combos"], ascending=[False, False])
             .head(30))
if not top_share.empty:
    plt.figure(figsize=(10, 5))
    plt.bar(top_share["query"], top_share["share_faster"])
    plt.xticks(rotation=60, ha="right")
    plt.ylabel("Share of combinations faster than default")
    plt.title("Top-30 Queries by Share of Faster Combinations")
    plt.tight_layout()
    plt.show()

# -------- 图3：组合数最多的前10个查询的速度提升比箱线图 --------
top_by_count = (summary.sort_values("n_combos", ascending=False)
                       .head(10)["query"].tolist())
box_data = [all_data.loc[all_data["query"] == q, "speedup"].values for q in top_by_count]
labels = top_by_count
if len(box_data) > 0:
    plt.figure(figsize=(10, 5))
    plt.boxplot(box_data, labels=labels, showfliers=False)
    plt.axhline(1.0, linestyle="--")
    plt.ylabel("Speedup (default / combo)")
    plt.title("Speedup Distribution for Top-10 (by #combos) Queries")
    plt.xticks(rotation=60, ha="right")
    plt.tight_layout()
    plt.show()

# -------- 打印几个关键信息表 --------
print("\n=== 更快组合占比 TOP 10 ===")
print(summary.sort_values(["share_faster", "n_combos"], ascending=[False, False])
            .head(10)
            .loc[:, ["query","n_combos","default_ms","share_faster","best_ms","best_speedup","median_speedup"]])

print("\n=== 最佳速度提升 TOP 10 ===")
print(summary.sort_values(["best_speedup", "n_combos"], ascending=[False, False])
            .head(10)
            .loc[:, ["query","n_combos","default_ms","best_ms","best_speedup","share_faster"]])
```

```python
# ========= 显著性与鲁棒性分析 =========
import math
REL_DELTAS = [0.00, 0.02, 0.05, 0.10, 0.20]  # 相对阈值：0/2/5/10/20%
ABS_MS = 50.0                                # 绝对阈值：至少快 50ms（按你集群噪声可改）

def is_sig_faster(default_ms, combo_ms, rel_delta, abs_ms=ABS_MS):
    return (default_ms - combo_ms) >= max(abs_ms, default_ms * rel_delta)

# 每查询：在不同阈值下的“显著更快占比”
sig_rows = []
for q, g in all_data.groupby("query"):
    d = float(g["default_ms"].iloc[0])
    row = {"query": q, "n_combos": len(g), "default_ms": d}
    for rd in REL_DELTAS:
        row[f"share_sig_{int(rd*100)}p"] = (g["combo_ms"].apply(lambda x: is_sig_faster(d, x, rd))).mean()
    sig_rows.append(row)
sig_df = pd.DataFrame(sig_rows)

# Wilson 区间
def wilson_ci(k, n, z=1.96):
    if n == 0: return (0.0, 0.0)
    phat = k / n
    denom = 1 + z**2/n
    center = (phat + z*z/(2*n)) / denom
    half = (z/denom)*math.sqrt(phat*(1-phat)/n + z*z/(4*n*n))
    return max(0.0, center-half), min(1.0, center+half)

def add_wilson_for(sig_df, col, prefix):
    lo, hi = [], []
    for _, r in sig_df.iterrows():
        n = int(r["n_combos"])
        k = int(round(r[col]*n))
        l,h = wilson_ci(k, n)
        lo.append(l); hi.append(h)
    sig_df[f"{prefix}_lo"] = lo
    sig_df[f"{prefix}_hi"] = hi

add_wilson_for(sig_df, "share_sig_5p",  "share_sig_5p")
add_wilson_for(sig_df, "share_sig_10p", "share_sig_10p")

# 全局层面（两种口径）
def global_combo_share(rel_delta):
    flags = all_data.apply(lambda r: is_sig_faster(r["default_ms"], r["combo_ms"], rel_delta), axis=1)
    n = len(flags); k = int(flags.sum())
    p = k / n if n > 0 else 0.0
    lo, hi = wilson_ci(k, n)
    return p, lo, hi, n

def bootstrap_ci(values, B=4000, alpha=0.05, rng=np.random.RandomState(42)):
    values = np.asarray(values, float)
    if values.size == 0: return (np.nan, np.nan, np.nan)
    boot = rng.choice(values, size=(B, values.size), replace=True).mean(axis=1)
    boot.sort()
    lo = np.percentile(boot, 100*alpha/2)
    hi = np.percentile(boot, 100*(1-alpha/2))
    return values.mean(), lo, hi

def global_query_share(sig_df, col):
    vals = sig_df[col].to_numpy()
    return bootstrap_ci(vals, B=4000)

print("\n===== 显著性总览（两种口径）=====")
for rd in REL_DELTAS:
    p_c, lo_c, hi_c, n_c = global_combo_share(rd)
    mean_q, lo_q, hi_q = global_query_share(sig_df, f"share_sig_{int(rd*100)}p")
    print(f"阈值≥{int(rd*100)}% & ≥{int(ABS_MS)}ms：")
    print(f"  [按组合加权]  占比={p_c:.2%} (95%CI {lo_c:.2%}~{hi_c:.2%}), 组合数={n_c}")
    print(f"  [按查询等权]  占比={mean_q:.2%} (95%CI {lo_q:.2%}~{hi_q:.2%}), 查询数={sig_df.shape[0]}")

# 符号检验：有多少“查询内显著更快占比>50%”
def exact_binom_two_sided(k, n, p0=0.5):
    from math import comb
    probs = [comb(n, i)*(p0**i)*((1-p0)**(n-i)) for i in range(n+1)]
    pk = probs[k]
    return sum(p for p in probs if p <= pk)

for rd in [0.05, 0.10]:
    col = f"share_sig_{int(rd*100)}p"
    flags = (sig_df[col] > 0.5).astype(int).values
    k = int(flags.sum()); n = int(flags.size)
    pval = exact_binom_two_sided(k, n, p0=0.5) if n <= 800 else np.nan
    print(f"\n[符号检验] 阈值≥{int(rd*100)}%：{k}/{n} 个查询中“显著更快占比”>50%。"
          + ("" if np.isnan(pval) else f" p-value={pval:.4g}"))

# 占比-阈值曲线
xs = [int(rd*100) for rd in REL_DELTAS]
combo_line, query_line = [], []
for rd in REL_DELTAS:
    combo_line.append(global_combo_share(rd)[0])
    query_line.append(global_query_share(sig_df, f"share_sig_{int(rd*100)}p")[0])

plt.figure(figsize=(7,5))
plt.plot(xs, combo_line, marker="o", label="按组合加权")
plt.plot(xs, query_line, marker="o", label="按查询等权")
plt.xlabel("显著更快的相对阈值（%）"); plt.ylabel("占比（更快）")
plt.title(f"显著更快占比 vs 阈值（绝对阈值≥{int(ABS_MS)}ms）")
plt.grid(True); plt.legend(); plt.tight_layout(); plt.show()

# 森林图（Top-25 by n_combos）
topN = sig_df.sort_values("n_combos", ascending=False).head(25)
ypos = np.arange(len(topN))
plt.figure(figsize=(8,8))
plt.hlines(ypos, topN["share_sig_5p_lo"], topN["share_sig_5p_hi"])
plt.plot(topN["share_sig_5p"], ypos, "o", label="点估计")
plt.axvline(0.5, linestyle="--", color="gray", label="0.5")
plt.yticks(ypos, topN["query"])
plt.xlabel("查询内 '显著更快(≥5%且≥50ms)' 的占比（95% Wilson CI）")
plt.title("按组合数 Top-25 查询的显著更快占比区间")
plt.legend(); plt.tight_layout(); plt.show()
```

```python
# verify_cbo.py
import os, glob
import numpy as np
import pandas as pd

ORIGIN = "Origin/origin.csv"   # 第一列：q开头查询名；第二列：原生 spark 运行时间（无表头）
ALL_DIR = "All"                # 里面每个 qX.csv 无表头；第一列为空，第二列是各组合的运行时间

def load_origin(path):
    df = pd.read_csv(path, header=None, usecols=[0, 1], names=["query", "baseline_ms"])
    df["query"] = df["query"].astype(str).str.strip()
    df["baseline_ms"] = pd.to_numeric(df["baseline_ms"], errors="coerce")
    df = df[np.isfinite(df["baseline_ms"]) & (df["baseline_ms"] > 0)]
    return df

def find_csv_for_query(q):
    # 优先 q.csv / q.CSV；找不到就用通配符
    candidates = [
        os.path.join(ALL_DIR, f"{q}.csv"),
        os.path.join(ALL_DIR, f"{q}.CSV"),
    ]
    for p in candidates:
        if os.path.exists(p): return p
    for pat in [os.path.join(ALL_DIR, f"{q}*.csv"),
                os.path.join(ALL_DIR, f"{q}*.CSV")]:
        got = glob.glob(pat)
        if got: return got[0]
    return None

def read_times_second_col(path):
    # 只读第二列；无表头；过滤非数值/非正数
    col = pd.read_csv(path, header=None, usecols=[1]).iloc[:, 0]
    col = pd.to_numeric(col, errors="coerce")
    col = col[np.isfinite(col) & (col > 0)]
    return col

def main():
    origin = load_origin(ORIGIN)

    rows, missing = [], []
    for _, r in origin.iterrows():
        q = r["query"]; base = float(r["baseline_ms"])
        f = find_csv_for_query(q)
        if not f:
            missing.append(q)
            rows.append({
                "query": q, "baseline_ms": base,
                "min_ms": np.nan, "max_ms": np.nan,
                "faster": False, "speedup": np.nan,
                "best_row": np.nan, "worst_row": np.nan,
                "file": None,
                "regret_spark": np.nan,          # 新增
                "regret_pct": np.nan             # 新增（百分比）
            })
            continue

        times = read_times_second_col(f)
        if len(times) == 0:
            rows.append({
                "query": q, "baseline_ms": base,
                "min_ms": np.nan, "max_ms": np.nan,
                "faster": False, "speedup": np.nan,
                "best_row": np.nan, "worst_row": np.nan,
                "file": f,
                "regret_spark": np.nan,          # 新增
                "regret_pct": np.nan             # 新增
            })
            continue

        # 最快 / 最慢
        min_idx = int(times.idxmin())
        min_ms  = float(times.loc[min_idx])
        max_idx = int(times.idxmax())
        max_ms  = float(times.loc[max_idx])

        faster  = (min_ms < base)
        speedup = (base / min_ms) if np.isfinite(min_ms) and min_ms > 0 else np.nan

        # regret（原生 Spark 相对“最优组合”的倍数；=1 表示刚好最优）
        regret = (base / min_ms) if (np.isfinite(min_ms) and min_ms > 0) else np.nan
        regret_pct = (regret - 1.0) * 100.0 if np.isfinite(regret) else np.nan

        rows.append({
            "query": q,
            "baseline_ms": base,
            "min_ms": min_ms,
            "max_ms": max_ms,
            "faster": faster,
            "speedup": speedup,
            "best_row": min_idx + 1,   # 1-based
            "worst_row": max_idx + 1,  # 1-based
            "file": f,
            "regret_spark": regret,    # 新增
            "regret_pct": regret_pct   # 新增
        })

    report = pd.DataFrame(rows)
    # 排序优先更大提速/更小 regret
    report = report.sort_values(
        by=["faster", "regret_spark", "speedup"],
        ascending=[False, True, False]
    )
    report.to_csv("report.csv", index=False)

    # 统计汇总（只对匹配到 All/* 的样本）
    matched = report[report["min_ms"].notna()].copy()
    num_total   = len(origin)
    num_found   = len(matched)
    num_improve = int(matched["faster"].sum())
    ratio       = (num_improve / num_found) if num_found else 0.0

    print(f"Total queries: {num_total}")
    print(f"Successfully matched the number of queries for All/* files: {num_found}")
    print(f"There are faster combinations of queries: {num_improve} (proportion {ratio:.1%})")

    # regret 汇总（Spark baseline 相对最优组合）
    valid_regret = matched["regret_spark"].replace([np.inf, -np.inf], np.nan).dropna()
    if len(valid_regret) > 0:
        med_rg = float(np.median(valid_regret))
        mean_rg = float(np.mean(valid_regret))
        succ10 = float(np.mean(valid_regret <= 1.10))
        succ20 = float(np.mean(valid_regret <= 1.20))
        print(f"\n[Baseline vs Best] Regret(median)={med_rg:.3f}×  Regret(mean)={mean_rg:.3f}×")
        print(f"[Baseline vs Best] Success@10%={succ10:.1%}  Success@20%={succ20:.1%}")
    else:
        print("\n[Baseline vs Best] No valid regret data.")

    # 展示若干最显著改进的样例
    improved = report[report["faster"]].head(10)
    print("\nSample (the 10 most obvious accelerations):")
    for _, r in improved.iterrows():
        print(f"- {r['query']}: baseline={r['baseline_ms']:.0f} ms, "
              f"best={r['min_ms']:.0f} ms, speedup={r['speedup']:.2f}×, "
              f"regret={r['regret_spark']:.3f}×, "
              f"file={os.path.basename(r['file']) if pd.notna(r['file']) else 'NA'}, "
              f"best_row={int(r['best_row']) if pd.notna(r['best_row']) else -1}")

if __name__ == "__main__":
    main()
```

```python
def plot_from_report():
    import pandas as pd, matplotlib.pyplot as plt, re
    from math import ceil

    report = pd.read_csv("report.csv")

    # 自然排序：q1, q2, ..., q10
    def natkey(s):
        return tuple(int(x) if x.isdigit() else x
                     for x in re.findall(r'\d+|\D+', str(s)))

    plot_df = report.sort_values("query", key=lambda s: s.map(natkey)).reset_index(drop=True)

    # 按数量一分为二（奇数时前一张多一个）
    n = len(plot_df)
    mid = ceil(n / 2)

    def make_plot(df_slice, outfile, title):
        import matplotlib.pyplot as plt
        L = len(df_slice)
        plt.figure(figsize=(min(max(10, L * 0.35), 32), 6))
        plt.plot(df_slice["query"], df_slice["baseline_ms"],
                 "-o", label="Baseline (origin)", color="blue", linewidth=1.5, markersize=3)
        plt.plot(df_slice["query"], df_slice["min_ms"],
                 "-o", label="Fastest combo (All)", color="red", linewidth=1.5, markersize=3)
        
        plt.xticks(rotation=90)
        plt.xlabel("Query"); plt.ylabel("Time (ms)")
        plt.title(title)
        plt.legend()
        plt.tight_layout()
        plt.savefig(outfile, dpi=200)
        print(f"Saved line chart -> {outfile}")

    # 前一半
    make_plot(plot_df.iloc[:mid], "cbo_compare_part1.png", "Baseline vs Fastest Combo (Part 1/2)")
    # 后一半
    make_plot(plot_df.iloc[mid:], "cbo_compare_part2.png", "Baseline vs Fastest Combo (Part 2/2)")

if __name__ == "__main__":
    main()
    plot_from_report()
```
